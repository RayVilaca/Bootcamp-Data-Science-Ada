### E ordenar a query

- ORDER BY

- ASC -> crescente;
- DESC -> decrescente.


```python
query = """
SELECT
    id_aluno,
    cpf AS CPF,
    primeiro_nome AS "Primeiro nome",
    ultimo_nome AS "Último nome",
    telefone_a AS "Primeiro telefone"
FROM
alunos
order by id_aluno asc


LIMIT 5
"""

ps.sqldf(query)
```

### Filtragem de colunas

```python
#onde id_aluno maior igual a 5

query = """
SELECT
    id_aluno,
    cpf AS CPF,
    primeiro_nome AS "Primeiro nome",
    ultimo_nome AS "Último nome",
    telefone_a AS "Primeiro telefone"
FROM
alunos
where id_aluno = 5

"""

ps.sqldf(query)
```

```python
# onde id_aluno menor que 4
query = """
SELECT
    id_aluno,
    cpf AS CPF,
    primeiro_nome AS "Primeiro nome",
    ultimo_nome AS "Último nome",
    telefone_a AS "Primeiro telefone"
FROM
alunos
WHERE id_aluno <= 4

"""

ps.sqldf(query)
```

### Exercício 1:

### Exercício 4:

Query que retorne apenas alunos que tem 2 telefones;

```python
query = """
SELECT
  id_aluno,
  cpf,
  primeiro_nome,
  telefone_a,
  telefone_b
FROM alunos
WHERE telefone_a != 0 and telefone_b != 0
"""

ps.sqldf(query)
```

### Exercício 5:

Query que retorne apenas alunos que tem 1 telefone apenas;

```python
query = """
SELECT
  id_aluno,
  cpf,
  primeiro_nome,
  telefone_a,
  telefone_b
FROM alunos
WHERE (telefone_a = 0 and telefone_b != 0) or (telefone_a != 0 and telefone_b = 0)
"""

ps.sqldf(query)
```

### Exercício 6:

Query que retorne apenas alunos com 1 telefone que moram em SP (2);

```python
# id_cidade = 2

query = """
SELECT
  id_aluno,
  cpf,
  primeiro_nome,
  id_cidade,
  telefone_a,
  telefone_b
FROM alunos
WHERE ((telefone_a = 0 and telefone_b != 0) or (telefone_a != 0 and telefone_b = 0)) and id_cidade = 2
"""

ps.sqldf(query)
```

### Exercício 14:

Query que retorne apenas alunos com primeira letra do nome e do sobrenome sendo vogais;

```python
query = """
SELECT *
FROM alunos
WHERE
  (primeiro_nome like 'a%' or primeiro_nome like 'e%' or primeiro_nome like 'i%' or primeiro_nome like 'o%' or primeiro_nome like 'u%') and
  (ultimo_nome like 'a%' or ultimo_nome like 'e%' or ultimo_nome like 'i%' or ultimo_nome like 'o%' or ultimo_nome like 'u%')
"""

ps.sqldf(query)
```
